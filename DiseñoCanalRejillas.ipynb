```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

Qmax = 0.4282 #Caudal, m^3/s
Vai=0.4 # Límite inferior de la velocidad de aproximación, m/s
Vas=0.9 # Límite superior de la velocidad de aproximación, m/s
S= 0.0001 # Pendiente del cana, m/m
hb= 0.6 # Bordo libre del canal, m
W = 1.60 # Ancho del canal, m
n= 0.012 # Coeficiente de rugosidad, [-]
```

```python
def RH(h):
    return W*h/(2*h+W)

def v(h):    
    vel = (1/n)*(RH(h))**(2/3.0)*S**(1/2.0)    
    return vel
    
def Q(h):
    caudal=v(h)*W*h   
    return caudal
```

```python
h=np.arange(0.0,1.0,0.01)
```

```python
plt.plot(h,RH(h),label="$RH(h)$")
plt.plot(h,v(h),label="v(h)")
plt.plot(h,Q(h), label="Q(h)")
plt.xlabel("Tirante hidraulico")
plt.axhline(0.4282, color="k")
plt.legend()
plt.show()
```

```python
plt.plot(h,Q(h))
plt.axhline(0.4282, color="k")
plt.legend()
plt.show()
```

```python
EPSILON=0.0001
for i in h:
    error=np.abs(Q(i)- Qmax)
    if error <= EPSILON:
        print (i)
        break
        
```
