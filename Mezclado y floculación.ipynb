# Mezclado y floculacin

El mezclado es una operacin unitaria que se utiliza para:
* Mezclar dos o mas sustancias
* Meclar liquidos miscibles
* Flocular partculas en el agua residual
* Mezclado continuo de suspensiones
* Transferir calor

El mezclado se clasifica en rpido continuo $ <30\,s $ y en continuo.


## Mezclado rpido continuo

Se utiliza para:

* Mezclar sustancias con el agua residual

> * Adicin de sales de aluminio o hierro para la floculacin
> * Dispersin de cloro para la desinfeccin en el agua tratada

* Mezclar lquidos miscibles
* Adicionar de qumicos a los lodos/bioslidos

## Mezclado continuo

Se utiliza en:

* Tanques de homogenizacin para mantener los slidos(material) en suspensin.
* Tanques de floculacin
* Reactores de lodos activados 
* Lagunas aireadas
* Digestores aerbicos


## Floculacin

Objetivo: Formar agregados o floculos a partir de partculas finamente divididas, o bien de partculas quimicamente desestabilizadas.

1. La floculacin incrementa la remocin de slidos suspendidos y de DBO en sedimentadores primarios.
2. Acondiciona el agua residual que contienen residuos industriales.
3. Mejora el desempeo de los tanques de sedimentacin secundarios en el proceso de lodos activados.
4. Sirve como pretratamiento para la filtracin

Existen dos tipos de floculacin:




1. Microfloculacin (floculacin perecintica): agreacin de partculas debido al movimiento aleatorio de las molculas del fluido.

> 1. Movimiento browniano
> 2. $0.001,\,a\, 1\mu m$

2. Macrofloculacin (Floculacin ortocintica)

> 1. Agregacin de partculas de $1,\, a\, 2\mu m$
> 2. Se da por gradientes de velocidad o sedimentacin diferencial



Camp y Stein(1943) estudiaron el efecto del gradiente de velocidad en tanques de coagulacin, llegaron a la siguiente ecuacin:

$$G=\sqrt{\frac{P}{\mu\, V}} $$

donde:
* $G=$ gradiente de velocidad promedio en el fluido, $s^{-1}$
> * Valores  grandes de G  se observa cerca de las paletas del dispositivo de mezclado
* $P=$ potencia requerida, $W$
* $\mu=$ viscosidad dinmica, $N\cdot s\, m^{-2}$
* $V=$ volumen del floculador, $m^3$

Multiplicado $G\times \frac{V}{Q}$

$$G_{\tau}=\frac{V}{Q}\sqrt{\frac{P}{\mu\, V}}=\frac{1}{Q}\sqrt{\frac{P\,V}{\mu}} $$ 

donde:
* $\tau=$ tiempo de retencin, $S$
* $Q=$ caudal, $m^3\,s^{-1}$


El concepto del gradiente de velocidad $G$ no aplica para microfloculacin, ni se puede usar como parmetro de diseo para algunos dispositivos de mezclado. Kolmogoroff describe el tamao de remolinos formados como resultado de la potencia suministrada al fluido.

$$I_{K}=\left(\frac{v^3}{P_M}\right)^{\frac{1}{4}} $$

donde:
* $I_K=$ longitud de Kolmogoroff(microescala), $m$
* $v= $ viscosidad cinemtica, $m^2\,s^{-1}$
* $P_M= G^2\,v$ potencia por unidad de masa, $W\,kg^{-1},\left[(kg\cdot m^2/s^3) kg^{-1} \right]$

$$I_{K}=\left(\frac{v^2}{G^2}\right)^{\frac{1}{4}} $$

* Si $G=1000\,s^{-1}$ y $v=1.00\times 10^{-6}\,m^2\,s^{-1}$ a $20^o\, C$, entonces $I_K=31.7\,\mu m$. Por lo tanto, partculas menores a $31.7\,\mu m$ no se veran afectadas.
* Si $G=10000\, s^{-1}$, la longitud de microescala es de $10\,\mu m$, por lo tanto si desea remover partculas ms pequeas que  $10\,\mu m$ se tienen que desestabilizar.

La efectividad del mezclado depende de:
* Caractersticas de la bomba
* Geometra del mezclador

## Tipos de mezcladores

### Estticos
* in-line static mixer with internal vanes
* in-line static mixer with orifice for mixing dilute chemicals
* in-line  mixer
* in-line mixer with internal mixer
* high-speed induction mixer
* pressurized water jet mixer with reactor tube

Para mezcladores estticos, el grado de mezclado est relacionado a la prdida de carga(cada de presin) en el mezclador:

$$h\approx k\left(\frac{v^2}{2\,g} \right)\approx K_{SM}v^2 $$

donde:
* $h = $ prdida de carga, $m$
* $k = $ coeficiente emprico caracterstico del mezclador
* $v = $ velocidad de aproximacin, $m\, s^{-1}$
* $K_{SM}= $ coeficiente total para el dispositico, $s^{2}\, m^{-1}$
> * $K_{SM}=(1.0,4.0)$ siendo 2.5 valor tpico para dispositivos usados en el tratamiento de agua
* $g=9.81\,m\, s^{-2}$

La potencia disipada por mezcladores estticos se obtiene por:

$$P=\gamma\,Q\,h $$
donde:
* $P=$ potencia disipada, $kW$
* $\gamma =$ peso especfico del agua, $kN\, m^{-3}$
* $Q=$ caudal, $m^3\,s^{-1}$

### Turbinas y propelas
Se usan para:
* Mantener material en suspensin
* Aireacin

Mezclado rpido ocurre en regimen turbulento (fuerzas inerciales predominan).
Como regla general, a altas velocidades y turbulencia, es ms eficiente el mezclado.

La siguiente expresin matemtica se puede usar para estimar los requerimientos de potencia y capacidad de bombeo:

$$P=N_{p}\,\rho\,n^3\,D^5$$

$$Q_i=N_{Q}\,n\,D^3$$

donde:

* $P=$ potencia requerida, $W$
* $N_{P}=$ Nmero de potencia para el impulsor, $[-]$
* $\rho=$ densidad, $kg\,m^{-3}$
* $n= $ revoluciones por segundo, $r\,s^{-1}$
* $D=$ Dimtro del impulsor, $m$
* $Q_i=$ Descarga, $m^3\,s^{-1}$
* $N_Q=$ Nmero de flujo para el impulsor, $[-]$

La ecuacin $P=N_{p}\,\rho\,n^3\,D^5$ aplica solamente si el rgimen es turbulento $(Re=10000$):

$$Re=\frac{D^2\,n\,\rho}{\mu}$$

donde :

* $n=$ velocidad rotacional o revoluciones por segundo, $r\,s^{-1}$
* $D,\,\rho,\,\mu$ ya se han definido previamente

La seleccin del mezclado se hace con base a pruebas de laboratorio o planta piloto, o bien por datos que proporciona la empresa proveedora del equipo.

Cuando se utilizan turbinas o propelas, es necesario evitar que se formen remolinos. Para ello es nesario colocar el equipo fuera del centro del tanque o a cierto ngulo de la vertical. En tanques rectangulares o circulares, se colocan cuatro o ms deflectores verticales(bafles) (Fig. 5-14).

Consideraciones de diseo:
* Gradiente de velocidad
* Velocidad rotacioneal o revoluciones por segundo
* Razn del dimetro del impulsor a el dimetro equivalente del tanque

## Tipos de mezcladores usados para el mezclado continuo en tratamiento de agua

1. Mezcladores neumticos
> Se inyecta aire u oxgeno en el fondo del tanque, la turbulencia causada por el burbujeo sirve para mezclar el contenido del tanque.

> En aireacin, las burbujas deben tener un tamao de $5\, mm$ de dimetro, y el flujo de aire debe ser de aproximadamente $10\,\%$ del caudal. El gradiente de velocidad $(G)$ formado por el burbujeo debe estar en el intervalo de $G_{avg}<200\,s^{-1}\,a\, G_{max}=8200\,s^{-1} $

> Cuando se inyecta aire para mezclar, o en tanques de floculacin o canales, la potencia disipada por el burbujeo se puede calcular con la siguiente ecuacin:
$$P=p_{a}V_{a}\ln\frac{p_c}{p_a} $$

> donde:
> * $P=$ potencia disipada, $W$
> * $p_a=$ presin atmsferica, $kN\,m^{-2}$
> * $V_a=$ flujo de aire a presin atmferica, $m^3\,s^{-1}$
> * $p_c=$ presin del aire en el punto de descarga, $kN\,m^{-2}$

$$p_c=K\,Q_a\ln\left(\frac{h+10.33}{10.33} \right)$$

> donde:
> * $K=1.689$
> * $Q_a=$ flujo de aire a presin atmsferica, $m^3\,min^{-1}$
> * $h=$ presin del aire en el punto de descarga, expresada en metros de agua, $m$
