# Mezclado y floculación

El mezclado es una operación unitaria que se utiliza para:
* Mezclar dos o mas sustancias
* Meclar liquidos miscibles
* Flocular partículas en el agua residual
* Mezclado continuo de suspensiones
* Transferir calor

El mezclado se clasifica en rápido continuo $ <30\,s $ y en continuo.


## Mezclado rápido continuo

Se utiliza para:

* Mezclar sustancias con el agua residual

> * Adición de sales de aluminio o hierro para la floculación
> * Dispersión de cloro para la desinfección en el agua tratada

* Mezclar líquidos miscibles
* Adicionar de químicos a los lodos/biosólidos

## Mezclado continuo

Se utiliza en:

* Tanques de homogenización para mantener los sólidos(material) en suspensión.
* Tanques de floculación
* Reactores de lodos activados 
* Lagunas aireadas
* Digestores aeróbicos


## Floculación

Objetivo: Formar agregados o floculos a partir de partículas finamente divididas, o bien de partículas quimicamente desestabilizadas.

1. La floculación incrementa la remoción de sólidos suspendidos y de DBO en sedimentadores primarios.
2. Acondiciona el agua residual que contienen residuos industriales.
3. Mejora el desempeño de los tanques de sedimentación secundarios en el proceso de lodos activados.
4. Sirve como pretratamiento para la filtración

Existen dos tipos de floculación:




1. Microfloculación (floculación perecinética): agreación de partículas debido al movimiento aleatorio de las moléculas del fluido.

> 1. Movimiento browniano
> 2. $0.001,\,a\, 1\mu m$

2. Macrofloculación (Floculación ortocinética)

> 1. Agregación de partículas de $1,\, a\, 2\mu m$
> 2. Se da por gradientes de velocidad o sedimentación diferencial



Camp y Stein(1943) estudiaron el efecto del gradiente de velocidad en tanques de coagulación, llegaron a la siguiente ecuación:

$$G=\sqrt{\frac{P}{\mu\, V}} $$

donde:
* $G=$ gradiente de velocidad promedio en el fluido, $s^{-1}$
> * Valores  grandes de G  se observa cerca de las paletas del dispositivo de mezclado
* $P=$ potencia requerida, $W$
* $\mu=$ viscosidad dinámica, $N\cdot s\, m^{-2}$
* $V=$ volumen del floculador, $m^3$

Multiplicado $G\times \frac{V}{Q}$

$$G_{\tau}=\frac{V}{Q}\sqrt{\frac{P}{\mu\, V}}=\frac{1}{Q}\sqrt{\frac{P\,V}{\mu}} $$ 

donde:
* $\tau=$ tiempo de retención, $S$
* $Q=$ caudal, $m^3\,s^{-1}$


El concepto del gradiente de velocidad $G$ no aplica para microfloculación, ni se puede usar como parámetro de diseño para algunos dispositivos de mezclado. Kolmogoroff describe el tamaño de remolinos formados como resultado de la potencia suministrada al fluido.

$$I_{K}=\left(\frac{v^3}{P_M}\right)^{\frac{1}{4}} $$

donde:
* $I_K=$ longitud de Kolmogoroff(microescala), $m$
* $v= $ viscosidad cinemética, $m^2\,s^{-1}$
* $P_M= G^2\,v$ potencia por unidad de masa, $W\,kg^{-1},\left[(kg\cdot m^2/s^3) kg^{-1} \right]$

$$I_{K}=\left(\frac{v^2}{G^2}\right)^{\frac{1}{4}} $$

* Si $G=1000\,s^{-1}$ y $v=1.00\times 10^{-6}\,m^2\,s^{-1}$ a $20^o\, C$, entonces $I_K=31.7\,\mu m$. Por lo tanto, partículas menores a $31.7\,\mu m$ no se veran afectadas.
* Si $G=10000\, s^{-1}$, la longitud de microescala es de $10\,\mu m$, por lo tanto si desea remover partículas más pequeñas que  $10\,\mu m$ se tienen que desestabilizar.

La efectividad del mezclado depende de:
* Características de la bomba
* Geometría del mezclador

## Tipos de mezcladores

### Estáticos
* in-line static mixer with internal vanes
* in-line static mixer with orifice for mixing dilute chemicals
* in-line  mixer
* in-line mixer with internal mixer
* high-speed induction mixer
* pressurized water jet mixer with reactor tube

Para mezcladores estáticos, el grado de mezclado está relacionado a la pérdida de carga(caída de presión) en el mezclador:

$$h\approx k\left(\frac{v^2}{2\,g} \right)\approx K_{SM}v^2 $$

donde:
* $h = $ pérdida de carga, $m$
* $k = $ coeficiente empírico característico del mezclador
* $v = $ velocidad de aproximación, $m\, s^{-1}$
* $K_{SM}= $ coeficiente total para el dispositico, $s^{2}\, m^{-1}$
> * $K_{SM}=(1.0,4.0)$ siendo 2.5 valor típico para dispositivos usados en el tratamiento de agua
* $g=9.81\,m\, s^{-2}$

La potencia disipada por mezcladores estáticos se obtiene por:

$$P=\gamma\,Q\,h $$
donde:
* $P=$ potencia disipada, $kW$
* $\gamma =$ peso específico del agua, $kN\, m^{-3}$
* $Q=$ caudal, $m^3\,s^{-1}$

### En linea
Son similares a los estáticos, pero contienen elementos moviles que mejoran el proceso de mezclado.

### Turbinas y propelas
Se usan para:
* Mantener material en suspensión
* Aireación

Mezclado rápido ocurre en regimen turbulento (fuerzas inerciales predominan).
Como regla general, a altas velocidades y turbulencia, es más eficiente el mezclado.

La siguiente expresión matemática se puede usar para estimar los requerimientos de potencia y capacidad de bombeo:

$$P=N_{p}\,\rho\,n^3\,D^5$$

$$Q_i=N_{Q}\,n\,D^3$$

donde:

* $P=$ potencia requerida, $W$
* $N_{P}=$ Número de potencia para el impulsor, $[-]$
* $\rho=$ densidad, $kg\,m^{-3}$
* $n= $ revoluciones por segundo, $r\,s^{-1}$
* $D=$ Diámtro del impulsor, $m$
* $Q_i=$ Descarga, $m^3\,s^{-1}$
* $N_Q=$ Número de flujo para el impulsor, $[-]$

La ecuación $P=N_{p}\,\rho\,n^3\,D^5$ aplica solamente si el régimen es turbulento $(Re=10000$):

$$Re=\frac{D^2\,n\,\rho}{\mu}$$

donde :

* $n=$ velocidad rotacional o revoluciones por segundo, $r\,s^{-1}$
* $D,\,\rho,\,\mu$ ya se han definido previamente

La selección del mezclado se hace con base a pruebas de laboratorio o planta piloto, o bien por datos que proporciona la empresa proveedora del equipo.

Cuando se utilizan turbinas o propelas, es necesario evitar que se formen remolinos. Para ello es nesario colocar el equipo fuera del centro del tanque o a cierto ángulo de la vertical. En tanques rectangulares o circulares, se colocan cuatro o más deflectores verticales(bafles) (Fig. 5-14).

Consideraciones de diseño:
* Gradiente de velocidad
* Velocidad rotacioneal o revoluciones por segundo
* Razón del diámetro del impulsor a el diámetro equivalente del tanque

## Tipos de mezcladores usados para el mezclado continuo en tratamiento de agua

1. Mezcladores neumáticos
> Se inyecta aire u oxígeno en el fondo del tanque, la turbulencia causada por el burbujeo sirve para mezclar el contenido del tanque.

> En aireación, las burbujas deben tener un tamaño de $5\, mm$ de diámetro, y el flujo de aire debe ser de aproximadamente $10\,\%$ del caudal. El gradiente de velocidad $(G)$ formado por el burbujeo debe estar en el intervalo de $G_{avg}<200\,s^{-1}\,a\, G_{max}=8200\,s^{-1} $

> Cuando se inyecta aire para mezclar, o en tanques de floculación o canales, la potencia disipada por el burbujeo se puede calcular con la siguiente ecuación:
$$P=p_{a}V_{a}\ln\frac{p_c}{p_a} $$

> donde:
> * $P=$ potencia disipada, $W$
> * $p_a=$ presión atmósferica, $kN\,m^{-2}$
> * $V_a=$ flujo de aire a presión atmóferica, $m^3\,s^{-1}$
> * $p_c=$ presión del aire en el punto de descarga, $kN\,m^{-2}$

$$p_c=K\,Q_a\ln\left(\frac{h+10.33}{10.33} \right)$$

> donde:
> * $K=1.689$
> * $Q_a=$ flujo de aire a presión atmósferica, $m^3\,min^{-1}$
> * $h=$ presión del aire en el punto de descarga, expresada en metros de agua, $m$
